```python
import csv
f = open ('전국 박물관 유물정보(2018.4.20.).csv')
data = csv.reader(f)
next(data)
data = list(data)

name = input('보고싶은 유물의 시대를 입력해주세요. :')
cnt = {}
for row in data :
    if name in row[6] :
        if row[1] in cnt :
            cnt[row[1]] += 1
        else :
            cnt[row[1]] = 1

import matplotlib.pyplot as plt
name = []
result = []
dcnt = [i for i in cnt.items()]
dcnt.sort(key=lambda x: -x[1])
print(dcnt[:3])

f.close()    
```
